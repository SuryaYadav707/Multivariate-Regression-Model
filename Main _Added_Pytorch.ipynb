## Multivariate Regression 

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

class MultivariateRegression:
    def __init__(self):
        self.theta = None
        self.mean = None
        self.std = None

    def normalize(self, X):
        X = X.astype(float)
        X_normalized = (X - self.mean) / self.std
        return X_normalized

    def hypothesis(self, X):
        return np.dot(X, self.theta)

    def cost_function(self, X, y):
        m = len(y)
        h = self.hypothesis(X)
        J = (1 / (2 * m)) * np.sum((h - y) ** 2)
        return J

    def gradient_descent(self, X, y, learning_rate, iterations):
        m, n = X.shape
        self.theta = np.zeros((n, y.shape[1]))
        cost_history = []
        

        for _ in range(iterations):
            h = self.hypothesis(X)
            gradient = (1 / m) * np.dot(X.T, (h - y))
            self.theta -= learning_rate * gradient
            cost = self.cost_function(X, y)
            cost_history.append(cost)

        return cost_history

    def train(self, X, y, learning_rate=0.01, iterations=1000):
        self.mean = np.mean(X, axis=0)
        self.std = np.std(X, axis=0)
        X_normalized = self.normalize(X)
        X_normalized = np.c_[np.ones(X_normalized.shape[0]), X_normalized]  # Add bias term
        y = y.values.reshape(-1, y.shape[1])  # Ensure y is a numpy array
        cost_history = self.gradient_descent(X_normalized, y, learning_rate, iterations)
        return cost_history

    def predict(self, X):
        X_normalized = self.normalize(X)
        X_normalized = np.c_[np.ones(X_normalized.shape[0]), X_normalized]  # Add bias term
        return self.hypothesis(X_normalized)

def estimate(model, X_test, y_test):
    # Make predictions
    predictions = model.predict(X_test)
    
    # Calculate R-squared
    y_mean = np.mean(y_test)
    ss_total = np.sum((y_test - y_mean) ** 2)
    ss_residual = np.sum((y_test - predictions) ** 2)
    r_squared = 1 - (ss_residual / ss_total)
    
    # Calculate Mean Squared Error (MSE)
    mse = np.mean((predictions - y_test) ** 2)
    
    # Calculate Root Mean Squared Error (RMSE)
    rmse = np.sqrt(mse)
    
    # Print evaluation metrics
    np.set_printoptions(suppress=True)
    print(bold+"R-squared:\n"+reset, r_squared*100)
    print(bold+"Mean Squared Error (MSE):\n"+reset, mse)
    print(bold+"Root Mean Squared Error (RMSE):\n"+reset, rmse)
    return r_squared

if __name__ == "__main__":
    # Load training data
    train_data = pd.read_csv('newnewMPR6.csv')
    
    X = train_data.drop(['Price_in_thousands','Fuel_efficiency'], axis=1)
    y = train_data[['Price_in_thousands','Fuel_efficiency']]
    
   
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=45)

    # Instantiate and train the model
    model = MultivariateRegression()
    cost_history = model.train(X_train, y_train)
    np.set_printoptions(suppress=True)

    estimate(model, X_test, y_test)
    
    predictions = model.predict(X_test)
   
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

class MultivariateRegression:
    def __init__(self):
        self.theta = None
        self.mean = None
        self.std = None

    def normalize(self, X):
        X = X.astype(float)
        X_normalized = (X - self.mean) / self.std
        return X_normalized

    def hypothesis(self, X):
        return np.dot(X, self.theta)

    def cost_function(self, X, y, lambda_l1=0, lambda_l2=0):
        m = len(y)
        h = self.hypothesis(X)
        J = (1 / (2 * m)) * np.sum((h - y) ** 2)
        l1_term = lambda_l1 * np.sum(np.abs(self.theta))
        l2_term = lambda_l2 * np.sum(self.theta ** 2)
        J += (l1_term + l2_term) / (2 * m)
        return J

    def gradient_descent(self, X, y, learning_rate, iterations, lambda_l1=0, lambda_l2=0):
        m, n = X.shape
        self.theta = np.zeros((n, y.shape[1]))
        cost_history = []

        for epoch in range(iterations):
            h = self.hypothesis(X)
            gradient = (1 / m) * np.dot(X.T, (h - y))
            l1_reg = lambda_l1 * np.sign(self.theta)
            l2_reg = lambda_l2 * self.theta
            self.theta -= learning_rate * (gradient + l1_reg + l2_reg)
            cost = self.cost_function(X, y, lambda_l1, lambda_l2)
            cost_history.append(cost)
            if epoch > 0:
                print("Epoch:", epoch, "Cost:", cost_history[-1], "Difference:", abs(cost_history[-1] - cost_history[-2]))

        return cost_history

    def train(self, X, y, learning_rate=0.01, iterations=1000, lambda_l1=0, lambda_l2=0):
        self.mean = np.mean(X, axis=0)
        self.std = np.std(X, axis=0)
        X_normalized = self.normalize(X)
        X_normalized = np.c_[np.ones(X_normalized.shape[0]), X_normalized]  # Add bias term
        y = y.values.reshape(-1, y.shape[1])  # Ensure y is a numpy array
        cost_history = self.gradient_descent(X_normalized, y, learning_rate, iterations, lambda_l1, lambda_l2)
        return cost_history

    def predict(self, X):
        X_normalized = self.normalize(X)
        X_normalized = np.c_[np.ones(X_normalized.shape[0]), X_normalized]  # Add bias term
        return self.hypothesis(X_normalized)

def estimate(model, X_test, y_test):
    # Make predictions
    predictions = model.predict(X_test)
    
    # Calculate R-squared
    y_mean = np.mean(y_test)
    ss_total = np.sum((y_test - y_mean) ** 2)
    ss_residual = np.sum((y_test - predictions) ** 2)
    r_squared = 1 - (ss_residual / ss_total)
    
    # Calculate Mean Squared Error (MSE)
    mse = np.mean((predictions - y_test) ** 2)
    
    # Calculate Root Mean Squared Error (RMSE)
    rmse = np.sqrt(mse)
    
    # Print evaluation metrics
    np.set_printoptions(suppress=True)
    print("R-squared:", r_squared*100)
    print("Mean Squared Error (MSE):", mse)
    print("Root Mean Squared Error (RMSE):", rmse)
    return r_squared

if __name__ == "__main__":
    # Load training data
    train_data = pd.read_csv('MPR6.csv')
    
    X = train_data.drop(['Price_in_thousands','Fuel_capacity'], axis=1)
    y = train_data[['Price_in_thousands','Fuel_capacity']]
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=45)

    # Instantiate and train the model
    model = MultivariateRegression()
    cost_history = model.train(X_train, y_train, lambda_l1=0.01, lambda_l2=0.01)  # Set lambda_l1 and lambda_l2 for L1 and L2 regularization
    np.set_printoptions(suppress=True)

    estimate(model, X_test, y_test)
    
    predictions = model.predict(X_test)
```
