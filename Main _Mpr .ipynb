```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

class MultivariateRegression:
    def __init__(self):
        self.theta = None
        self.mean = None
        self.std = None

    def normalize(self, X):
        X = X.astype(float)
        X_normalized = (X - self.mean) / self.std
        return X_normalized

    def hypothesis(self, X):
        return np.dot(X, self.theta)

    def cost_function(self, X, y):
        m = len(y)
        h = self.hypothesis(X)
        J = (1 / (2 * m)) * np.sum((h - y) ** 2)
        return J

    def gradient_descent(self, X, y, learning_rate, iterations):
        m, n = X.shape
        self.theta = np.zeros((n, y.shape[1]))
        cost_history = []
        

        for _ in range(iterations):
            h = self.hypothesis(X)
            gradient = (1 / m) * np.dot(X.T, (h - y))
            self.theta -= learning_rate * gradient
            cost = self.cost_function(X, y)
            cost_history.append(cost)

        return cost_history

    def train(self, X, y, learning_rate=0.01, iterations=1000):
        self.mean = np.mean(X, axis=0)
        self.std = np.std(X, axis=0)
        X_normalized = self.normalize(X)
        X_normalized = np.c_[np.ones(X_normalized.shape[0]), X_normalized]  # Add bias term
        y = y.values.reshape(-1, y.shape[1])  # Ensure y is a numpy array
        cost_history = self.gradient_descent(X_normalized, y, learning_rate, iterations)
        return cost_history

    def predict(self, X):
        X_normalized = self.normalize(X)
        X_normalized = np.c_[np.ones(X_normalized.shape[0]), X_normalized]  # Add bias term
        return self.hypothesis(X_normalized)

def estimate(model, X_test, y_test):
    # Make predictions
    predictions = model.predict(X_test)
    
    # Calculate R-squared
    y_mean = np.mean(y_test)
    ss_total = np.sum((y_test - y_mean) ** 2)
    ss_residual = np.sum((y_test - predictions) ** 2)
    r_squared = 1 - (ss_residual / ss_total)
    
    # Calculate Mean Squared Error (MSE)
    mse = np.mean((predictions - y_test) ** 2)
    
    # Calculate Root Mean Squared Error (RMSE)
    rmse = np.sqrt(mse)
    
    # Print evaluation metrics
    np.set_printoptions(suppress=True)
    print("R-squared:", r_squared*100)
    print("Mean Squared Error (MSE):", mse)
    print("Root Mean Squared Error (RMSE):", rmse)
    return r_squared

if __name__ == "__main__":
    # Load training data
    train_data = pd.read_csv('MPR6.csv')
    
    X = train_data.drop(['Price_in_thousands','Horsepower'], axis=1)
    y = train_data[['Price_in_thousands','Horsepower']]
    
   
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=45)

    # Instantiate and train the model
    model = MultivariateRegression()
    cost_history = model.train(X_train, y_train)
    np.set_printoptions(suppress=True)

    estimate(model, X_test, y_test)
    
    predictions = model.predict(X_test)
   
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

class MultivariateRegression:
    def __init__(self):
        self.theta = None
        self.mean = None
        self.std = None

    def normalize(self, X):
        X = X.astype(float)
        X_normalized = (X - self.mean) / self.std
        return X_normalized

    def hypothesis(self, X):
        return np.dot(X, self.theta)

    def cost_function(self, X, y):
        m = len(y)
        h = self.hypothesis(X)
        J = (1 / (2 * m)) * np.sum((h - y) ** 2)
        return J

    def gradient_descent(self, X, y, learning_rate, iterations):
        m, n = X.shape
        self.theta = np.zeros((n, y.shape[1]))
        cost_history = []
        

        for _ in range(iterations):
            h = self.hypothesis(X)
            gradient = (1 / m) * np.dot(X.T, (h - y))
            self.theta -= learning_rate * gradient
            cost = self.cost_function(X, y)
            cost_history.append(cost)

        return cost_history

    def train(self, X, y, learning_rate=0.01, iterations=1000):
        self.mean = np.mean(X, axis=0)
        self.std = np.std(X, axis=0)
        X_normalized = self.normalize(X)
        X_normalized = np.c_[np.ones(X_normalized.shape[0]), X_normalized]  # Add bias term
        y = y.values.reshape(-1, y.shape[1])  # Ensure y is a numpy array
        cost_history = self.gradient_descent(X_normalized, y, learning_rate, iterations)
        return cost_history

    def predict(self, X):
        X_normalized = self.normalize(X)
        X_normalized = np.c_[np.ones(X_normalized.shape[0]), X_normalized]  # Add bias term
        return self.hypothesis(X_normalized)

def estimate(model, X_test, y_test):
    # Make predictions
    predictions = model.predict(X_test)
    
    # Calculate R-squared
    y_mean = np.mean(y_test)
    ss_total = np.sum((y_test - y_mean) ** 2)
    ss_residual = np.sum((y_test - predictions) ** 2)
    r_squared = 1 - (ss_residual / ss_total)
    
    # Calculate Mean Squared Error (MSE)
    mse = np.mean((predictions - y_test) ** 2)
    
    # Calculate Root Mean Squared Error (RMSE)
    rmse = np.sqrt(mse)
    
    # Print evaluation metrics
    np.set_printoptions(suppress=True)
    print("R-squared:", r_squared*100)
    print("Mean Squared Error (MSE):", mse)
    print("Root Mean Squared Error (RMSE):", rmse)
    return r_squared

if __name__ == "__main__":
    # Load training data
    train_data = pd.read_csv('synthetic_data.csv')
    
    X = train_data.drop(['Price_in_thousands','Horsepower'], axis=1)
    y = train_data[['Price_in_thousands','Horsepower']]
    
   
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=45)

    # Instantiate and train the model
    model = MultivariateRegression()
    cost_history = model.train(X_train, y_train)
    np.set_printoptions(suppress=True)

    estimate(model, X_test, y_test)
    
    predictions = model.predict(X_test)
   
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import pandas as pd
from sklearn.metrics import mean_squared_error, mean_absolute_error

data = pd.read_csv('MPR6.csv')

X = data.drop(['Price_in_thousands', 'Horsepower'], axis=1)
y = data[['Price_in_thousands', 'Horsepower']]

X_mean = X.mean()
X_std = X.std()
X_normalized = (X - X_mean) / X_std

X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.2, random_state=42)

X_train_tensor = torch.tensor(X_train.values, dtype=torch.float32)
X_test_tensor = torch.tensor(X_test.values, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train.values, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test.values, dtype=torch.float32)

class MultivariateRegression(nn.Module):
    def __init__(self, input_size, output_size):
        super(MultivariateRegression, self).__init__()
        self.linear = nn.Linear(input_size, output_size)

    def forward(self, x):
        return self.linear(x)

input_size = X_train_tensor.shape[1]  # Number of features
output_size = y_train_tensor.shape[1]  # Number of targets

# Initialize the model
model = MultivariateRegression(input_size, output_size)

# Define loss function and optimizer
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)

# Train the model
num_epochs = 1000
for epoch in range(num_epochs):
    # Forward pass
    outputs = model(X_train_tensor)
    loss = criterion(outputs, y_train_tensor)
    
    # Backward pass and optimization
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    if (epoch+1) % 100 == 0:
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')

# Test the trained model
with torch.no_grad():
    predicted = model(X_test_tensor)
    

    
    for i, col in enumerate(y.columns):
        y_true = y_test[col].values
        y_pred = predicted[:, i].numpy()
        mse = mean_squared_error(y_true, y_pred)
        rmse = mean_squared_error(y_true, y_pred, squared=False)
        mae = mean_absolute_error(y_true, y_pred)
        r2=r2_score(y_true, y_pred)
        
        print(f'Evaluation metrics for {col}:')
        print(f'Mean Squared Error (MSE): {mse:.4f}')
        print(f'Root Mean Squared Error (RMSE): {rmse:.4f}')
        print(f'Mean Absolute Error (MAE): {mae:.4f}')
        print(f'R2 Accuracy (R2): {r2:.4f}')

    print(predicted)
```
